```python
import numpy as np
import pandas as pd
import pylab as plt
```

```python
#写入数据
dta = np.array([4.101,3.297,3.533,5.687,6.778,4.873,3.592,3.973,2.731,3.557,2.863,4.170,
                4.225,2.581,1.965,4.257,4.373,3.573,3.320,2.257,3.110,4.574,5.328,2.645,
                2.859,3.721,3.836,2.417,3.074,3.483,3.847,3.250,3.735,4.842,3.564,3.109,
                2.463,1.778,1.450,1.956,2.196,4.584,3.715,1.853,2.543,2.123,2.756,3.690])
data = pd.Series(dta)
```

```python
# 画时间序列线图
data.plot(figsize=[12,4])
```

```python
from statsmodels.stats.diagnostic import unitroot_adf # 平稳性检验
from statsmodels.stats.diagnostic import acorr_ljungbox    #白噪声检验
```

```python
print(unitroot_adf(data))
print("-------+-+-+-+这-里-是-分-割-线+-+-+-+-----------")
print(acorr_ljungbox(data,lags=6))
```

```python
# 画自相关图和偏自相关图
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_acf(data.values.squeeze(),lags=20,ax=ax1)
fig = sm.graphics.tsa.plot_pacf(data,lags=20,ax=ax2)
```

```python
# 建模，并输出模型的参数
arma_mod = sm.tsa.ARMA(data,(0,1)).fit()
print(arma_mod.params)
```

```python
# 根据bic准则编写函数来推荐最优的模型：
def SelectModel(data):
    best_model = sm.tsa.ARMA(data,(0,0)).fit()
    for i in range(2):
        for j in range(2):
            arma_mod = sm.tsa.ARMA(data,(i,j)).fit()
            print('当p和q为，相应的bic值为：',i,j,arma_mod.bic)
            if best_model.bic > arma_mod.bic :
                best_model = arma_mod
    print('根据bic准则，选出最优的模型为：',best_model.summary2())
            
SelectModel(data)
```

```python
# 模型的检验
# 原假设是，序列剩下的没被用来预测的部分的信息为白噪声
# 若不拒绝原假设，则通过检验，模型显著
from scipy import stats
resid = arma_mod.resid
stats.normaltest(resid)
```
